```python
import pandas as pd
from pycaret.classification import *

# Load data
data = pd.read_excel("Dataset TWS.xlsx")

# Initialize setup
clf = setup(data, target='Program Studi')

# Compare models
best_model = compare_models()

# Create a Random Forest model
rf_model = create_model('dt')

# Tune the model
tuned_rf_model = tune_model(rf_model)

# Evaluate the model
evaluate_model(tuned_rf_model)

# Finalize the model
final_rf_model = finalize_model(tuned_rf_model)

# Make predictions
new_data = pd.DataFrame({
    "Jenjang Pendidikan": ["D3"],
    "Minat dan Bakat": ["Fisika, Teknik Tenaga Listrik (TTL)"],
    "Jalur Pendaftaran PENS": ["SBMPN (Seleksi Bersama Masuk Politeknik Negeri)"],
    "Rata-rata Nilai Masuk PENS": [474.0],
    "Rencana Karir": ["Network Engineer"]
})
predictions = predict_model(final_rf_model, data=new_data)

# Save the model
# save_model(final_rf_model, 'final_rf_model')
```
